## FINAL CODE EXERCISE 2

```python
#Import libraries to use

import bz2
import pandas as pd

## Chunk process. iterations in different dataframe
chunk_bookings = pd.read_csv('bookings.csv', sep='^', usecols =['arr_port', 'pax', 'year'], chunksize=1000000)
sum_chunks = pd.DataFrame()


for i, bookings in enumerate (chunk_bookings):
    print ("Chunk: %d, size of chunk %d"%(i+1, len(bookings)))
    bookings = bookings[bookings['year'] == 2013]
    bookings.drop ('year', axis =1, inplace=True)
    arrival_airports = bookings.groupby ('arr_port')
    chunk_result = arrival_airports.sum()
    sum_chunks = sum_chunks.append (chunk_result)
top10airports = sum_chunks.groupby(sum_chunks.index).pax.sum().sort_values(ascending = False).reset_index()

topairports = top10airports[0:10]



## Adding Airport name & city name

from neobase import NeoBase
geoDict = NeoBase()
geoDict.get (top10airports['arr_port'][0].strip())['name']
topairports['Airport_name'] = topairports['arr_port'].map(lambda x: geoDict.get (x.strip())['name'])
topairports['City_name'] = topairports['arr_port'].map(lambda x: geoDict.get (x.strip())['city_name_list'])

print (topairports)

## Output file

topairports.to_csv ('top_10_airports.csv', sep = '^', index = False)
print('File created succesfully in current path')
```
